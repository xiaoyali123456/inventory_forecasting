```python
import json
def parse(segs):
    default = ''
    try:
        js = json.loads(segs)
        if isinstance(js, list):
            lst = js
        else:
            lst = js.get('data', [])
    except:
        return default
    filter_ = set()
    for x in lst:
        for y in ['CITY_', 'STATE_']:
            if y in x:
                filter_.add(x)
    return '|'.join(filter_)

t['geo_segs'] = t.user_segments.map(parse)
t
```
