```python
join.to_csv('join.csv', index=False)
```

```python
err = join.groupby(['cd', 'y', 'y_hat']).agg(reach=('reach_y', 'sum'), reach_hat=('reach_x', 'sum'), num=('reach_y', 'size')).reset_index()
err['err'] = err.num / err.groupby(['cd', 'y'])['num'].transform('sum')
err.to_csv('err.csv', index=False)
```
